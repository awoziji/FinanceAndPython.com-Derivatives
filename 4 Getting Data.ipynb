```python
import quandl
data = quandl.get("CME/SPH2017")
data
```

```python
data["Settle"]
```

```python
data["Settle"].plot(kind="line")
plt.show()
```

```python
import pandas_datareader.data as web
import datetime
sp500 = web.DataReader("SP500", "fred", data.index[0], data.index[-1])
sp500
```

```python
df = pd.concat([data["Settle"],sp500],axis=1)

df.plot(kind="line")
plt.legend(["Futures Contract","SP500"])
plt.show()
```

```python
last = df.index[-1]
print(last)
df["T"] = (last-df.index).days/365
df["r"] = (df["Settle"]/df["SP500"])**-df["T"]-1
df["r"].plot(kind="line")
plt.ylabel("Implied r")
plt.xlabel("Date")
plt.title("S&P 500 Futures Contract r")
plt.show()
```

```python
years = ["2014","2015","2016"]
months = ["H","M","U","Z"]
symbols = []
for x in years:
    for y in months:
        symbols.append("CME/SP"+y+x)
data = quandl.get(symbols)
print(data)
```

```python
cols = [x+" - Settle" for x in symbols]
data = data[cols]
```

```python
data.plot(kind="line")
plt.legend(symbols,bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
sp500 = web.DataReader("SP500", "fred", data.index[0], data.index[-1])
```

```python
def impliedR(x):
    last = x.last_valid_index()
    t = pd.Series((last-x.index).days/365)
    t.index = x.index
    return (x/sp500["SP500"])**-t-1

rates = data.apply(impliedR)
rates.plot(kind="line")
plt.legend(symbols,bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```
